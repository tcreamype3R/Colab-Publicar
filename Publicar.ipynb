```python
# Script para an√°lisis de publicaciones de LinkedIn con mejor rendimiento
# Basado en el documento: Analisis (53%) Publicaciones (Ultimos 7 dias).docx

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de estilo para las visualizaciones
plt.style.use('default')
sns.set_palette("husl")

# Datos de las publicaciones (extra√≠dos del documento)
data = {
    'Ranking': [1, 2, 3, 4, 5],
    'Impresiones': [235, 118, 110, 108, 108],
    'Titulo': [
        'üéØ Presentando Proyecto de Tesis: News GoalTracker',
        'üîç News GoalTracker: Plataforma con 7 Agentes IA para investigaci√≥n',
        '¬°React no es solo una librer√≠a, es un ecosistema!',
        'FabLabArtDesign* Innovaci√≥n Artesanal con Tecnolog√≠a',
        'YUAN* China: IA desde los 6 a√±os y Yuan Digital'
    ],
    'Engagement': [8, 1, 0, 1, 0],
    'Dias_desde_publicacion': [1, 1, 6, 3, 4]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Mostrar tabla de publicaciones
print("üìä TOP 5 PUBLICACIONES POR IMPRESIONES (√öLTIMOS 7 D√çAS)")
print("=" * 85)
print(f"{'#':<3} {'Impresiones':<12} {'T√≠tulo/Tema':<50} {'Engagement':<12} {'D√≠as desde pub.':<15}")
print("-" * 85)
for _, row in df.iterrows():
    print(f"{row['Ranking']:<3} {row['Impresiones']:<12} {row['Titulo'][:47]:<50} {str(row['Engagement'])+' comentarios':<12} {row['Dias_desde_publicacion']:<15}")
print("=" * 85)

# An√°lisis de patrones de √©xito
print("\n\nüß† PATRONES DE √âXITO IDENTIFICADOS")
patrones = [
    "1. Tesis y Proyectos Personales: La publicaci√≥n sobre la tesis (News GoalTracker) obtuvo el mayor alcance (235 impresiones) y engagement (8 comentarios).",
    "2. Temas de Tecnolog√≠a y IA: Publicaciones sobre React, IA, Agentes IA, Yuan Digital tuvieron alto alcance.",
    "3. Uso de Emojis y Formato: Las publicaciones con emojis y formato dividido en secciones son m√°s atractivas.",
    "4. Hashtags Estrat√©gicos: Los m√°s efectivos son los espec√≠ficos como #NewsGoalTracker, #DataScience, #React, #Innovaci√≥n.",
    "5. Incluir Imagen o Miniatura de Enlace: Aumenta el CTR.",
    "6. Llamados a la Acci√≥n (CTA): Las publicaciones que invitan a comentar o participar tienen mejor engagement."
]

for patron in patrones:
    print(f"   ‚Ä¢ {patron}")

# Recomendaciones
print("\n\nüìå RECOMENDACIONES PARA FUTURAS PUBLICACIONES")
recomendaciones = [
    "‚úÖ Mantener el formato visual con emojis y bullets",
    "‚úÖ Contar una historia personal o de proyecto con lecciones aprendidas",
    "‚úÖ Usar entre 5 y 10 hashtags relevantes (no saturar)",
    "‚úÖ Incluir una imagen o miniatura para mejorar el atractivo visual",
    "‚úÖ Terminar con una pregunta o CTA para fomentar interacci√≥n",
    "‚úÖ Evitar muchos enlaces externos en una misma publicaci√≥n (puede penalizar el alcance)"
]

for rec in recomendaciones:
    print(f"   ‚Ä¢ {rec}")

# Visualizaciones
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gr√°fico de barras: Impresiones por publicaci√≥n
bars = ax1.bar(df['Ranking'].astype(str), df['Impresiones'], color=['#1a76d2', '#4285f4', '#6ba4f7', '#9dc3f7', '#cde1fb'])
ax1.set_title('Impresiones por Publicaci√≥n', fontsize=16, fontweight='bold')
ax1.set_xlabel('Ranking de Publicaci√≥n', fontsize=12)
ax1.set_ylabel('N√∫mero de Impresiones', fontsize=12)

# A√±adir etiquetas de valor en las barras
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 5,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# Gr√°fico de dispersi√≥n: Impresiones vs Engagement
scatter = ax2.scatter(df['Dias_desde_publicacion'], df['Impresiones'],
                     s=df['Engagement']*50, alpha=0.7, c=df['Ranking'], cmap='viridis')
ax2.set_title('Relaci√≥n: D√≠as desde publicaci√≥n vs Impresiones\n(Tama√±o = Engagement)', fontsize=16, fontweight='bold')
ax2.set_xlabel('D√≠as desde publicaci√≥n', fontsize=12)
ax2.set_ylabel('Impresiones', fontsize=12)
ax2.grid(True, alpha=0.3)

# A√±adir etiquetas a los puntos
for i, txt in enumerate(df['Ranking']):
    ax2.annotate(f'#{txt}', (df['Dias_desde_publicacion'][i], df['Impresiones'][i]),
                xytext=(5, 5), textcoords='offset points', fontweight='bold')

# A√±adir barra de color para el ranking
cbar = plt.colorbar(scatter, ax=ax2)
cbar.set_label('Ranking', fontsize=12)

plt.tight_layout()
plt.show()

# An√°lisis de correlaci√≥n
correlation = df[['Impresiones', 'Engagement', 'Dias_desde_publicacion']].corr()
print("\n\nüìà MATRIZ DE CORRELACI√ìN")
print("=" * 40)
print(correlation)

# Tendencias de contenido con potencial
print("\n\nüöÄ TENDENCIAS DE CONTENIDO CON POTENCIAL")
tendencias = [
    "‚Ä¢ Proyectos de IA aplicada (tesis, herramientas, agentes)",
    "‚Ä¢ Tecnolog√≠as frontend (React, ecosistemas de desarrollo)",
    "‚Ä¢ Innovaci√≥n educativa y metodolog√≠as √°giles",
    "‚Ä¢ Sostenibilidad y green tech",
    "‚Ä¢ An√°lisis de datos y casos de estudio"
]

for tendencia in tendencias:
    print(f"   {tendencia}")

# Conclusiones
print("\n\nüí° CONCLUSIONES")
print("   ‚Ä¢ Las publicaciones sobre proyectos personales y tesis generan mayor engagement")
print("   ‚Ä¢ El formato visual con emojis mejora el atractivo de las publicaciones")
print("   ‚Ä¢ Los temas de tecnolog√≠a y IA son los m√°s populares entre la audiencia")
print("   ‚Ä¢ Publicaciones m√°s recientes tienden a tener mayor alcance (sesgo temporal)")
```

```python
# Script para an√°lisis de publicaciones de LinkedIn con mejor rendimiento (14 d√≠as)
# Basado en el documento: Analisis (53%) Publicaciones (Ultimos 14 dias).docx

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuraci√≥n de estilo para las visualizaciones
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = [12, 8]

# Datos de las publicaciones (extra√≠dos del documento)
data = {
    'Publicacion': [
        'Reto100D√≠asDataScience',
        'Emprender Sin Miedo',
        'Timidez Femenina (Vulnerabilidad)',
        'Mentor-√çA (EdTech + IA)',
        'Presentaci√≥n Tesis (News GoalTracker)'
    ],
    'Impresiones': [518, 304, 259, 236, 235],
    'Reacciones': [5, 2, 3, 1, 1],
    'Comentarios': [1, 6, 2, 2, 8],
    'Compartidos': [2, 0, 0, 0, 1]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Calcular m√©tricas adicionales
df['Engagement_Total'] = df['Reacciones'] + df['Comentarios'] + df['Compartidos']
df['Tasa_Engagement'] = (df['Engagement_Total'] / df['Impresiones'] * 100).round(2)

# Mostrar tabla de publicaciones
print("üìä TOP 5 PUBLICACIONES POR IMPRESIONES (√öLTIMOS 14 D√çAS)")
print("=" * 95)
print(f"{'Publicaci√≥n':<40} {'Impresiones':<12} {'Reacciones':<10} {'Comentarios':<12} {'Compartidos':<10} {'Engagement':<10}")
print("-" * 95)
for _, row in df.iterrows():
    print(f"{row['Publicacion'][:38]:<40} {row['Impresiones']:<12} {row['Reacciones']:<10} {row['Comentarios']:<12} {row['Compartidos']:<10} {row['Engagement_Total']:<10}")
print("=" * 95)

# An√°lisis de tendencias e insights
print("\n\nüß† AN√ÅLISIS DE TENDENCIAS E INSIGHTS")

print("\n1. TEMAS CON MAYOR ENGAGEMENT:")
temas = [
    "‚Ä¢ Data Science y Retos educativos (518 impresiones) ‚Üí Alto inter√©s en formaci√≥n pr√°ctica y certificaci√≥n",
    "‚Ä¢ Emprendimiento y superaci√≥n personal (304 y 259 impresiones) ‚Üí Resonancia emocional y utilidad pr√°ctica",
    "‚Ä¢ Tecnolog√≠a aplicada (IA, EdTech, Tesis) ‚Üí Inter√©s en innovaci√≥n y casos de uso reales"
]
for tema in temas:
    print(f"   {tema}")

print("\n2. FORMATO Y ESTILO:")
formatos = [
    "‚Ä¢ Publicaciones con emojis, listas, preguntas y llamados a la acci√≥n (CTA) generan m√°s interacci√≥n",
    "‚Ä¢ Uso de hashtags estrat√©gicos y menciones a herramientas/plataformas (Google Colab, NotebookLM, GoalTracker)",
    "‚Ä¢ Inclusi√≥n de im√°genes o portadas de art√≠culos mejora el atractivo visual"
]
for formato in formatos:
    print(f"   {formato}")

print("\n3. TIPOS DE CONTENIDO QUE FUNCIONAN:")
tipos_contenido = [
    "üìö Gu√≠as descargables (PDFs, recursos gratuitos)",
    "üé§ Testimonios y casos de √©xito",
    "ü§ñ Temas de IA, automatizaci√≥n y ciencia de datos",
    "üí° Motivaci√≥n y desarrollo profesional",
    "üåç An√°lisis geopol√≠ticos y tecnol√≥gicos (Ej.: Yuan digital, Per√∫)"
]
for tipo in tipos_contenido:
    print(f"   ‚Ä¢ {tipo}")

print("\n4. PALABRAS CLAVE Y HASHTAGS M√ÅS USADOS:")
hashtags = ["#DataScience", "#Python", "#EmprenderSinMiedo", "#Innovaci√≥n", "#IA", "#GoalTracker", "#Tesis", "#TechLatam"]
print("   " + ", ".join(hashtags))

# Recomendaciones
print("\n\nüìå RECOMENDACIONES PARA FUTURAS PUBLICACIONES")
recomendaciones = [
    "‚úÖ Mantener el formato visual con emojis y bullets",
    "‚úÖ Incluir recursos descargables o enlaces a webinars",
    "‚úÖ Usar preguntas para fomentar comentarios",
    "‚úÖ Seguir explorando temas de IA, educaci√≥n y emprendimiento",
    "‚úÖ Publicar en d√≠as y horarios de alta actividad en LinkedIn (ma√±anas entre martes y jueves)"
]

for rec in recomendaciones:
    print(f"   ‚Ä¢ {rec}")

# Visualizaciones
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('AN√ÅLISIS DE RENDIMIENTO DE PUBLICACIONES (14 D√çAS)', fontsize=16, fontweight='bold')

# Gr√°fico 1: Impresiones por publicaci√≥n
bars1 = axes[0, 0].barh(df['Publicacion'], df['Impresiones'], color=['#1a76d2', '#4285f4', '#6ba4f7', '#9dc3f7', '#cde1fb'])
axes[0, 0].set_title('Impresiones por Publicaci√≥n', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('N√∫mero de Impresiones', fontsize=12)
for bar in bars1:
    width = bar.get_width()
    axes[0, 0].text(width + 5, bar.get_y() + bar.get_height()/2,
                   f'{int(width)}', ha='left', va='center', fontweight='bold')

# Gr√°fico 2: Engagement por tipo (reacciones, comentarios, compartidos)
engagement_data = df[['Reacciones', 'Comentarios', 'Compartidos']]
engagement_data.index = df['Publicacion']
engagement_data.plot(kind='bar', ax=axes[0, 1], stacked=True)
axes[0, 1].set_title('Desglose de Engagement por Publicaci√≥n', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Cantidad', fontsize=12)
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].legend(loc='upper right')

# Gr√°fico 3: Tasa de engagement
bars3 = axes[1, 0].barh(df['Publicacion'], df['Tasa_Engagement'], color=['#34a853', '#fbbc05', '#ea4335', '#4285f4', '#9aa0a6'])
axes[1, 0].set_title('Tasa de Engagement (%)', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Porcentaje de Engagement', fontsize=12)
for bar in bars3:
    width = bar.get_width()
    axes[1, 0].text(width + 0.05, bar.get_y() + bar.get_height()/2,
                   f'{width}%', ha='left', va='center', fontweight='bold')

# Gr√°fico 4: Relaci√≥n entre impresiones y engagement total
scatter = axes[1, 1].scatter(df['Impresiones'], df['Engagement_Total'], s=100, alpha=0.7)
axes[1, 1].set_title('Relaci√≥n: Impresiones vs Engagement Total', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Impresiones', fontsize=12)
axes[1, 1].set_ylabel('Engagement Total', fontsize=12)
axes[1, 1].grid(True, alpha=0.3)

# A√±adir etiquetas a los puntos
for i, txt in enumerate(df['Publicacion']):
    axes[1, 1].annotate(txt[:15] + "...", (df['Impresiones'][i], df['Engagement_Total'][i]),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

plt.tight_layout()
plt.show()

# An√°lisis de correlaci√≥n
correlation = df[['Impresiones', 'Reacciones', 'Comentarios', 'Compartidos', 'Engagement_Total']].corr()
print("\n\nüìà MATRIZ DE CORRELACI√ìN")
print("=" * 55)
print(correlation)

# Conclusiones
print("\n\nüí° CONCLUSIONES")
conclusiones = [
    "‚Ä¢ Las publicaciones con mayor rendimiento combinan valor educativo, relevancia tem√°tica y un llamado a la acci√≥n claro",
    "‚Ä¢ El contenido sobre Data Science, emprendimiento y tecnolog√≠a genera el mayor engagement",
    "‚Ä¢ Ofrecer recursos tangibles (PDF, webinar, herramienta) aumenta significativamente el engagement",
    "‚Ä¢ La tasa de engagement var√≠a considerablemente entre publicaciones, incluso con impresiones similares",
    "‚Ä¢ Los comentarios son el componente m√°s variable del engagement entre diferentes publicaciones"
]

for conclusion in conclusiones:
    print(f"   {conclusion}")

# M√©tricas resumen
print("\n\nüìã M√âTRICAS RESUMEN")
print(f"   ‚Ä¢ Total de impresiones: {df['Impresiones'].sum()}")
print(f"   ‚Ä¢ Promedio de impresiones por publicaci√≥n: {df['Impresiones'].mean():.0f}")
print(f"   ‚Ä¢ Total de engagement: {df['Engagement_Total'].sum()}")
print(f"   ‚Ä¢ Tasa de engagement promedio: {df['Tasa_Engagement'].mean():.2f}%")
print(f"   ‚Ä¢ Publicaci√≥n con mayor engagement: {df.loc[df['Engagement_Total'].idxmax()]['Publicacion']}")
print(f"   ‚Ä¢ Publicaci√≥n con mayor tasa de engagement: {df.loc[df['Tasa_Engagement'].idxmax()]['Publicacion']} ({df['Tasa_Engagement'].max()}%)")
```
